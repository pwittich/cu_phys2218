```python
import matplotlib.pyplot as plt
import numpy as np
import cmath
import math
from axes import phy_axes

%matplotlib inline
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 200  # 200 e.g. is really fine, but slower
```

```python
omega0=1
omegac=.5
lambdap = omega0
lambdam = math.sqrt(omega0**2 + omegac**2)

tvals = np.linspace(0, 150, 7500)
yvals1 = np.cos(lambdap*tvals)
yvals2 = np.cos(lambdam*tvals)
plt.plot(tvals, yvals1, label="$\lambda_-$")
plt.plot(tvals, yvals2, label="$\lambda_+$")
plt.title("Particle 1")
plt.grid()
plt.legend()
phy_axes()
plt.show()
```

```python
# beat behavior in 1st particle
plt.plot(tvals, yvals1-1.2*yvals2, label="$\lambda_-$+$\lambda_+$")
plt.grid()
plt.title("$\omega_0=%3.1f$ Hz, $\omega_c=%3.1f$ Hz, $\lambda_-$=%3.1f Hz, $\lambda_+$=%3.1f Hz" %
          (omega0, omegac, lambdap, lambdam))
plt.xlabel("time (s)")
plt.ylabel("Amplitude")
phy_axes()
plt.show()
```

```python
omega0=1
omegac=.5
lambdap = omega0
lambdam = math.sqrt(omega0**2 + omegac**2)

a1 = 1

x1 = a1*(np.cos(lambdap*tvals) + np.cos(lambdam*tvals))
x2 = a1*(np.cos(lambdap*tvals) - np.cos(lambdam*tvals))
fig,axes = plt.subplots(2,1)
axes[0].plot(tvals, x1)
axes[0].grid(True)
axes[0].set_xlabel("time (s)")
axes[0].set_ylabel("Amplitude $x_1$")
phy_axes(ax=axes[0])
axes[1].plot(tvals, x2)
#axes[1].grid(True)
axes[1].set_xlabel("time (s)")
axes[1].set_ylabel("Amplitude $x_2$")
phy_axes(ax=axes[1])
fig.suptitle(f"$\omega_0={omega0:.2f}$ Hz, $\omega_c={omegac:.1f}$ Hz, $\omega_-$={lambdam:.1f} Hz, $\omega_+$={lambdap:.1f} Hz")
plt.show()

plt.savefig("lec04_beats.png")
```
