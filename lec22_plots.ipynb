```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
# retina display
plt.rcParams['figure.dpi'] = 300
plt.rcParams['axes.grid'] = True
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
alpha = np.linspace(-3*np.pi, 3*np.pi, 2500)
plt.plot(alpha, np.sin(alpha)**2/alpha**2)
plt.grid(True)
plt.ylabel("$I_0$")
plt.xlabel(r"$\alpha$")
plt.title(r"Single slit diffraction pattern, $I_0$ vs $\alpha$")
plt.savefig("lec22_singeslit.png")
```

```python
from matplotlib.ticker import FuncFormatter, MultipleLocator
# now as a function of theta
theta = np.linspace(-np.pi/4, np.pi/4, 1000)
I_0=2
llambda = 1
k = 2 * np.pi / llambda
a = llambda*2
alpha = k*a*np.sin(theta)
fig, ax = plt.subplots()
ax.plot(theta, I_0*np.sin(alpha)**2/alpha**2, label=f"$\lambda={llambda}, a = {a}$")
ax.xaxis.set_major_formatter(FuncFormatter(
    lambda val, pos: '{:1.3g}$\pi$'.format(val/np.pi) if val != 0 else '0'
))
ax.xaxis.set_major_locator(MultipleLocator(base=np.pi/8))
plt.xlabel(r"$\theta$")
plt.ylabel(r"$\langle I(\theta)\rangle$")
plt.legend()
plt.savefig("lec22_singeslit_theta.png")
#plt.plot(theta, I_0*np.sin(alpha)**2/alpha**2)
```
