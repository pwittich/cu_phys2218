```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
# retina display
plt.rcParams['figure.dpi'] = 100

# Constants
wavelength = 650e-9  # Wavelength of the light (650 nm, red light)
slit_width = 2e-6  # Width of the slit (2 micrometers)

# Function to calculate intensity as a function of viewing angle


def diffraction_intensity(theta, wavelength, slit_width):
    beta = np.pi * slit_width * np.sin(theta) / wavelength
    intensity = (np.sin(beta) / beta)**2
    intensity[beta == 0] = 1  # Handle the division by zero case
    return intensity


# Viewing angles from -90 to 90 degrees
theta = np.linspace(-np.pi / 2, np.pi / 2, 1000)

# Calculate the diffraction pattern
intensity = diffraction_intensity(theta, wavelength, slit_width)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(theta * 180 / np.pi, intensity, label='Diffraction Pattern')
plt.axvline(0, color='red', linestyle='--', label='Center (θ = 0°)')
plt.xlabel('Viewing Angle θ (degrees)')
plt.ylabel('Normalized Intensity')
plt.title('Single Slit Diffraction Pattern')
plt.legend()
plt.grid(True)
# Add text box
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
textstr = r'$I(\theta) = I_0 \left(\frac{\sin(\alpha)}{\alpha}\right)^2$' + \
    '\n' + r'$\alpha \equiv \frac{\pi a \sin(\theta)}{\lambda}$'
plt.text(0.02, 0.98, textstr, transform=plt.gca().transAxes, fontsize=18,
         verticalalignment='top', bbox=props)

plt.savefig('diffraction_pattern.png')
```
