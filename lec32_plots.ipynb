```python
import matplotlib.pyplot as plt
import numpy as np
#from ugh import ugh

%matplotlib inline
# retina display
plt.rcParams['figure.dpi'] = 300
plt.rcParams['axes.grid'] = True
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
# n factorial
def n_factorial(n):
    if n == 0:
        return 1
    else:
        return n * n_factorial(n - 1)

def binominals(n, k):
    return n_factorial(n) / (n_factorial(k) * n_factorial(n - k))
```

```python
n=12
for k in range(0, n + 1):
    print(n, k, binominals(n, k))
```

```python
#plt.xkcd()


def addlabels(x, y):
    for i in range(len(x)):
        plt.text(i-0.25, y[i]+10, y[i])

vals = [binominals(n, k) for k in range(0, n + 1)]
plt.bar(range(0, n + 1), vals, width=1.0, edgecolor='black')
plt.xlabel('bin (number of right hops)')
plt.ylabel(r'binomial factor $\frac{n!}{m!(n-m)!}$')
addlabels(range(0, n + 1), vals)
plt.savefig('lec32_plinko_binomials.png')
```

```python
# n choose m
def n_choose_m(n, m):
    return n_factorial(n) / (n_factorial(m) * n_factorial(n - m))

print(3, 2, n_choose_m(3, 2))
```

```python
def einstein_solid(n, q):
    return n_choose_m(q+n-1, q)
```

```python
print(3, 2, einstein_solid(3, 2))
```
