```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
import math


# retina distplay
%matplotlib inline
# hidpi display
plt.rcParams['figure.dpi'] = 300
# turn on grid by default
plt.rcParams['axes.grid'] = True
```

```python

B = 1 # bulk modulus
t = 0
omega = 200*2*math.pi # in Hz
c = 331 # in m/s
k = omega / c # 1/m
llambda = 2*math.pi/k # in meters
print(llambda)

xvals = np.linspace(0, 3*llambda, 500)

amplitude = 1
xi = amplitude * np.cos(k*xvals - omega * t)
pressure = -k*B * np.sin(k*xvals - omega * t)

plt.plot(xvals, xi, label=r'displacement $\xi(x,t)$')
plt.plot(xvals, pressure, label=r'pressure difference $p(x,t)$')
plt.legend()
plt.xlabel(r'$x [m]$')
plt.ylabel('various')
plt.title(f'$\lambda$ = {llambda:.1f} m')
plt.savefig('lec15_xi_p.png')
```
