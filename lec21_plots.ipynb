```python
from matplotlib.ticker import FuncFormatter, MultipleLocator
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
from axes import axes
%matplotlib inline


# retina display
plt.rcParams['figure.dpi'] = 300
plt.rcParams['axes.grid'] = True
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
beta = np.linspace(-np.pi, np.pi, 2500)
I0=1
N=3
I=I0*np.sin(N*beta)**2/np.sin(beta)**2

fig, ax = plt.subplots()
ax.plot(beta, I, label=f"N={N}")
ax.xaxis.set_major_formatter(FuncFormatter(
    lambda val, pos: '{:.0g}$\pi$'.format(val/np.pi) if val != 0 else '0'
))
ax.xaxis.set_major_locator(MultipleLocator(base=np.pi/2))
plt.ylabel('$I(\\beta)/I_0$')
plt.legend()
axes()
plt.savefig(f"lec21_n{N}.png")
```

```python
theta = np.linspace(-np.pi,np.pi, 2500)
llambda = 5
k = 2*np.pi/llambda
b = llambda/4
N = 3
I = I0*(np.sin(N*k*b*np.sin(theta)/2)**2)/(np.sin(k*b*np.sin(theta)/2)**2)
plt.polar(theta, I, label=f"N={N}")
```

```python
#now as a function of theta
plt.rcParams['figure.figsize'] = [5, 5]

llambda=2
k=2*np.pi/llambda
b=0.5*llambda
nvals = [3,5,10, 15]
print (nvals)
theta=np.linspace(0.01, 2*np.pi, 2500)


for i,N in enumerate(nvals):
    I = I0*np.sin(N*k*b*np.sin(theta)/2)**2/np.sin(k*b*np.sin(theta)/2)**2
    ax= plt.subplot(2,2,i+1,projection='polar')
    ax.plot(theta, I, label=f"N={N}")
    plt.legend()
plt.savefig(f"lec21_polar_nmany.png")
```

```python
import ipywidgets as widgets

# use widget rather than inline plot
%matplotlib widget
plt.rcParams['figure.dpi'] = 100
plt.rcParams['axes.grid'] = True
plt.rcParams['figure.figsize'] = [10, 5]

# set up plot
#fig, ax = plt.subplots(1, 1, subplot_kw=dict(polar=True))
fig, ax = plt.subplots(subplot_kw = {'projection': 'polar'})
# generate theta values
theta = np.linspace(0.01, 2*np.pi, 2500)

llambda = 1 # with this choice, b is in fractions of lambda 
k = 2*np.pi/llambda

@widgets.interact(N=(0, 10, 1), b=(0, 2, .125))
def update(N=2,b=0.5):
    """Remove old lines from plot and plot new one"""
    [l.remove() for l in ax.lines]


    I = I0*np.sin(N*k*b*np.sin(theta)/2)**2/np.sin(k*b*np.sin(theta)/2)**2
    ax.plot(theta, I)
 
```
