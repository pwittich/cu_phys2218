```python
import scipy.special as sp
import matplotlib.pyplot as plt
import numpy as np
from axes import axes

%matplotlib inline
# retina display
plt.rcParams['figure.dpi'] = 100
plt.rcParams['axes.grid'] = True
```

```python
u = np.linspace(0, 3*np.pi, 1000)
intensity = sp.jv(1, u)**2/u**2

psi = sp.jv(1,u)/u
```

```python
def find_roots(x,y):
    s = np.abs(np.diff(np.sign(y))).astype(bool)
    return x[:-1][s] + np.diff(x)[s]/(np.abs(y[1:][s]/y[:-1][s])+1)
```

```python
plt.plot(u, psi)
plt.ylabel(r'$J_1(u)/u$')
plt.xlabel(r'$u$')
axes()
z=find_roots(u, psi)
print(z)
print(z[1]/np.pi)
plt.plot(z, np.zeros(len(z)), marker="o", ls="", ms=4)
plt.text(z[1]-0.1, .02, f'${z[1]/np.pi:1.3g}\pi$')
plt.text(z[2]-0.1, .02, f'${z[2]/np.pi:1.3g}\pi$')
plt.savefig("lec24_bessel1.png")
```

```python
# can't find the zeros here directly, since the algorithm looks for 0 crossing and this doesn't cross
plt.plot(u, intensity)
axes()
plt.xlabel(r'$u$')
plt.ylabel(r'$\left( J_1(u)/u \right)^2$')
plt.plot(z, np.zeros(len(z)),marker="o", ls="", ms=4)
plt.text(z[1]-0.1, .02, f'${z[1]/np.pi:1.3g}\pi$')
plt.text(z[2]-0.1, .02, f'${z[2]/np.pi:1.3g}\pi$')
plt.savefig("lec24_bessel2.png")
```
