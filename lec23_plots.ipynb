```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
# retina display
plt.rcParams['figure.dpi'] = 100
plt.rcParams['axes.grid'] = True
```

```python
theta = np.linspace(-np.pi/8, np.pi/8, 5000)

llambda = 800 # in nm
k = 2 * np.pi / llambda
a = llambda * 5
b = 2*a #1e3/(300.6e-3) # in nm
print(b)
N = 3
I_0 = 2

# multi slit diffraction pattern
def f(theta):
    alpha = k * a * np.sin(theta)/2
    beta = k * b * np.sin(theta)/2
    return I_0 * np.sin(alpha)**2/(alpha**2) * np.sin(N*beta)**2/np.sin(beta)**2

# multiple source interference pattern
def e1(theta):
    beta = k * b * np.sin(theta)/2
    return I_0*np.sin(N*beta)**2/np.sin(beta)**2

# single slit diffraction pattern
def e2(theta):
    alpha = k * a * np.sin(theta)/2
    return I_0 * np.sin(alpha)**2/(alpha**2) 
```

```python
from matplotlib.ticker import FuncFormatter, MultipleLocator
fig, ax = plt.subplots()

ax.plot(theta, f(theta), 'b', label=f'{N} slit diffraction pattern')
ax.plot(theta, e1(theta), 'r--', label=f'{N} source inteference pattern')
sf=1
ax.plot(theta, sf*e2(theta), 'g', label=f'single slit diffraction pattern')
ax.xaxis.set_major_formatter(FuncFormatter(
    lambda val, pos: '{:1.3g}$\pi$'.format(val/np.pi) if val != 0 else '0'
))
ax.xaxis.set_major_locator(MultipleLocator(base=np.pi/16))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel(r'$\theta$')
plt.ylabel('$I/I_0$')
plt.savefig(f'lec23_diffraction_pattern_N{N}.png', bbox_inches="tight")
```

```python
# interference pattern, renormalized to be at the same overall scale (divide by N^2)
theta = np.linspace(-0.15, 0.15, 5000)
N=2
n2 = e1(theta)/N**2
N=4
n4 = e1(theta)/N**2
N=10
n10 = e1(theta)/N**2
plt.plot(theta, n2, 'b', label=f'Two source interference pattern')
plt.plot(theta, n4, 'g', label=f'Four source interference pattern')
plt.plot(theta, n10, 'r', label=f'Ten source interference pattern')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel(r'$\theta$')
plt.ylabel('$I/(I_0N^2)$')
plt.savefig(f'lec23_interference_pattern_N2410.png', bbox_inches="tight")
```
