```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
import math
from axes import axes 

# retina distplay
%matplotlib inline
# hidpi display
plt.rcParams['figure.dpi'] = 300
# turn on grid by default
plt.rcParams['axes.grid'] = True
```

```python
xvals = np.linspace(0, 1, 500)
llambda = 1
k = 2*np.pi/llambda
period = 10
omega = 2*np.pi/period
tvals = np.linspace(0, period, 20)
t=0

def xi(x: np.ndarray, pmbar: int) -> np.ndarray:
    if pmbar == 0:
        val = -1
    else:
        val = +1
    return np.cos(k*x - val* omega*t)
```

```python
plt.plot(xvals, xi(xvals, 0))
```

```python
L = 1
def kval(n: np.ndarray) -> np.ndarray:
    return n*2*np.pi/L
# nvals is a list of ints from 1 to 5
nvals = np.ndarray([ i for i in range(6)])
print(nvals, kval(nvals))
#for k in kval(nvals):
#    plt.plot(xvals, np.cos(k*xvals))
```

```python
L = 1


# nvals is a list of ints from 1 to 5
nvals = np.ndarray([i for i in range(6)])
print(nvals, kval(nvals))
plt.subplot(3, 1, 1)
for n in [1,2,3]:
    plt.subplot(3, 1, n)
    kk = n*np.pi/(L)
    plt.plot(xvals, np.cos(kk*xvals), 'b', label=f'n={n}')
    plt.plot(xvals, -np.cos(kk*xvals), 'b')
    plt.legend()
    plt.ylabel(r'$\xi(x,t)$')
plt.suptitle(f'$\\xi(x,t)$ vs $x$ for free ends')
plt.xlabel(r'$x/L$')
plt.savefig('lec14_xi_freends_n123.png')
#axes()
```

```python


for n in [1, 3,  5]:
    kk = n*np.pi/(L)
    #print(kk, n, 2*np.pi/kk)
    plt.plot(xvals, np.cos(kk*xvals), label=f'n={n}')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title(f'$\\xi(x,t)$ vs $x$ for free ends ($n$ odd)')
axes()
plt.savefig('lec14_xi_freends_nodd.png', bbox_inches="tight")
```

```python


for n in [1,2,3, 4]:
    kk = (n-1/2)*np.pi/(L)
    #print(kk, n, 2*np.pi/kk)
    plt.plot(xvals, np.sin(kk*xvals), label=f'n={n}')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title(f'$\\xi(x,t)$ vs $x$ for one free, one fixed ends')
plt.xlabel(r'$x/L$')
axes()
plt.savefig('lec14_xi_onefreeonefixed_n1234.png', bbox_inches="tight")
```
